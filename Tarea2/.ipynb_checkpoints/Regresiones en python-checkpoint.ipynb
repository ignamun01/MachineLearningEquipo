```python
train, test = train_test_split(bd, test_size = 0.30, random_state = 100) 
```

```python
y = train.default
```

```python
#dtype="categorical", solver="mip",
   #                   cat_cutoff=0.1)
```

```python
optb_patrimonio = OptimalBinning(name="patrimonio", dtype="numerical", solver="cp")
optb_Destino = OptimalBinning(name="Destino", dtype="categorical", solver="mip",)
optb_Prest_Original = OptimalBinning(name="Prest_Original", dtype="numerical", solver="cp")
optb_R_Endeudamiento = OptimalBinning(name="R_Endeudamiento", dtype="numerical", solver="cp")
optb_R_Patrim_Activo = OptimalBinning(name="R_Patrim_Activo", dtype="numerical", solver="cp")
optb_bureau_score = OptimalBinning(name="bureau_score", dtype="numerical", solver="cp")
optb_genero = OptimalBinning(name="genero", dtype="categorical", solver="mip",)
optb_situ_laboral = OptimalBinning(name="situ_laboral", dtype="categorical", solver="mip",)
optb_activo = OptimalBinning(name="activo", dtype="numerical", solver="cp")
optb_pasivo = OptimalBinning(name="pasivo", dtype="numerical", solver="cp")
optb_ingreso = OptimalBinning(name="ingreso", dtype="numerical", solver="cp")
optb_edad = OptimalBinning(name="edad", dtype="numerical", solver="cp")
```

```python
optb_patrimonio.fit(train["patrimonio"].values, y)
optb_Destino.fit(train["Destino"].values, y)
optb_Prest_Original.fit(train["Prest_Original"].values, y)
optb_R_Endeudamiento.fit(train["R_Endeudamiento"].values, y)
optb_R_Patrim_Activo.fit(train["R_Patrim_Activo"].values, y)
optb_bureau_score.fit(train["bureau_score"].values, y)
optb_genero .fit(train["genero"].values, y)
optb_situ_laboral.fit(train["situ_laboral"].values, y)
optb_activo.fit(train["activo"].values, y)
optb_pasivo.fit(train["pasivo"].values, y)
optb_ingreso .fit(train["ingreso"].values, y)
optb_edad.fit(train["edad"].values, y)
```

```python
#Transformamos las variables a woe

train["woe_patrimonio"] = optb_patrimonio.transform(train["patrimonio"], metric="woe")
train["woe_Destino"] = optb_Destino.transform(train["Destino"], metric="woe")
train["woe_Prest_Original"] = optb_Prest_Original.transform(train["Prest_Original"], metric="woe")
train["woe_R_Endeudamiento"] = optb_R_Endeudamiento.transform(train["R_Endeudamiento"], metric="woe")
train["woe_R_Patrim_Activo"] = optb_R_Patrim_Activo.transform(train["R_Patrim_Activo"], metric="woe")
train["woe_bureau_score"] = optb_bureau_score.transform(train["bureau_score"], metric="woe")
train["woe_genero"] = optb_genero.transform(train["genero"], metric="woe")
train["woe_situ_laboral"] = optb_situ_laboral.transform(train["situ_laboral"], metric="woe")
train["woe_activo"] = optb_activo.transform(train["activo"], metric="woe")
train["woe_pasivo"] = optb_pasivo.transform(train["pasivo"], metric="woe")
train["woe_ingreso"] = optb_ingreso.transform(train["ingreso"], metric="woe")
train["woe_edad"] = optb_edad.transform(train["edad"], metric="woe")
```

```python
X_woe_train=train.iloc[:,13:24]
y_train=train.iloc[:,12]
X_woe_train.head()
```

```python
corr = train.iloc[:,1:].corr()
corr

xticks = list(corr.index)  
yticks = list(corr.index)  
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(1, 1, 1)

sns.heatmap(corr,annot=True,cmap="rainbow",ax=ax1,linewidths=.5,annot_kws={'size':9,'weight':'bold', 'color':'k'})
ax1.set_xticklabels(xticks, rotation=30, fontsize=10)
ax1.set_yticklabels(yticks, rotation=0, fontsize=10)
plt.show()
```

```python
model=LogisticRegression()
clf=model.fit(X_woe_train,y_train)
```

```python

y_pred=clf.predict(X_woe_train)
y_pred1=clf.decision_function(X_woe_train)
print('y_pred\n',y_pred)
print('y_pred1\n',y_pred1)

#Draw ROC curve and calculate AUC value
fpr,tpr,threshold = roc_curve(y_train, y_pred1)
roc_auc = auc(fpr,tpr)
plot(fpr,tpr,color='darkorange',
         label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC_curve')
plt.legend(loc="lower right")
plt.show()
```
